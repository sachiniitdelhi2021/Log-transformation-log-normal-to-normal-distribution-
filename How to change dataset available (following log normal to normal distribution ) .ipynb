```python
# for checking the given dataset follows Normal-distribution we use Q-Q plot(quantile-quantile plot) as 

# we will import all important required libraries like
import numpy as np      # for generating random data points 
import statsmodels.api as sm              # for ploting Q-Q plot  
import pylab as py    # for getting the plot 

#first we will generate random data points following normal distribution like
data_points= np.random.normal(0,1,100)

# plotting getting Q-Q plot using 
sm.qqplot(data_points,line='45')  # line='45' -- is just for the thickness of the line 
py.show()
```

```python
# in above plot we get one straight line which is generated  by the system itself which basically 
# represents the points generated by the system following normal distribution perfectly 
# this line is taken as reference to compare how much data given is close to follow normal distribution 
# every time we run above code we will get different plot boz the generated datapoints will change every time 
```

```python
# we can cross check this also by plotting the given datapopints like 
import seaborn as sns
sns.displot(data_points)     # we are getting histogram for given data like 
```

```python
# if we want to see the PDF(probability density function) we can get that as 
sns.displot(data_points,kind='kde')   
# kind='kde' -- for kernal density extimation for getting bell curve as 
# this confirms the data points follows normal distribution 
```

```python
# here we have the 100 random points which follows normal distribution 
data_points
```

```python
# now instead of generating data which follows  normal distribution we will generate data which log normal distribution 
# and we will see how it's Q-Q  plot look and then we will change it to nornal distribution like 
data_points2=np.random.lognormal(0,1,100)

# same as above case 
sm.qqplot(data_points2,line='45')
py.show()
```

```python
# see the data is how much away  from the reference line which means the data is not following normal distribution
```

```python
# how it's bell plot look like 
sns.displot(data_points2,kind='kde')
# we can see below the bell plot is not exactly bell 
# so we need to make it perfect like
```

```python
# using 'np.log(data_points2)'
# now we will change this log normal to normal as 

sm.qqplot(np.log(data_points2),line='45')
py.show()
```

```python
# now we can see the data is close to the reference line and is much more normally distributed as compare to previos time 
# again plotting it's bell curve as
sns.displot(np.log(data_points2),kind='kde')
# see below it is much improved compare to previous bell curve 
```
